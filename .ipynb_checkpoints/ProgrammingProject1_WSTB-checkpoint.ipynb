```python
from math import e
import decimal
getcontext().prec = 2

start=float(input("Enter the Starting Population: "))
rate=float(input("Enter the rate that the population is growing anually: "))
totalyears=int(input("Enter the total amount of years you would like to calculate for: "))
popdouble=start*2

for currentyear in range(totalyears):
    while start>=popdouble:
        print("")
        print ("During year", currentyear+1, "***Population has doubled***")
        print("")
        x=currentyear*rate
        end = start * e ** x
        print ("The current population is", end)
        print ("Estimated growth was", end - start)
        start=end
        popdouble=start*2
    else:
        print("")
        print ("During year", currentyear+1)
        print("")
        x=currentyear*rate
        end = start * e ** x
        print ("The current population is", end)
        print ("Estimated growth was", end - start)
        start=end
```
